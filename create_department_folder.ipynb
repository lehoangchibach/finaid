```python
import os
import shutil
import pandas as pd
import numpy as np
```

```python
# Import the Student Placement Master as csv file.
# Need to have columns: "Department", "R#", "First", "Last"
df = pd.read_csv("./data/Incoming Student Placement(Student Master List).csv") # Give the code a path to the csv here
df.head(1)
```

```python
# Just making sure "Department" column is not empty
df = df[df["Department"].notna()]
```

```python
# Just making sure "Department" column is not empty
df = df[df["Department"] != 0]
```

```python
# In "./data" folder, download all the incoming student application folders into "Incoming Students Apps". 
# This folder represents all the students that HAVE a folder.
rids = {}
student_dir = "./data/Incoming Students Apps/"
for x in os.listdir(student_dir):
    if "-" not in x:
        continue
    
    rid = x.rsplit("-", 1)[1]
    rids[rid.strip()] = os.path.join(student_dir, x)
```

```python
df.columns[:15]
```

```python
# Making sure every student in the Student Placement Master List 
# has their folder created in "./data/Incoming Students Apps/"
for i,data in df.iterrows():
    if data["R#"] not in rids:
        print(data)
    assert data["R#"] in rids
```

```python
# Create folders for each department
# Make sure you have a folder "All Departments" in the "./data" folder
dep_path = "./data/All Departments/"
dep_dirs = os.listdir(dep_path)
for i, data in df.iterrows():
    department = data["Department"].replace("/", "_").strip()
    if department not in dep_dirs:
        os.makedirs(os.path.join(dep_path, department))
```

```python
# Main loop
# For each row (student) in Placement Masterlist, 
# copy that student folder to the corresponding department folder.
for i, data in df.iterrows():
    department = data["Department"].replace("/", "_").strip()
    
    fold_name = f"{data["First"]} {data["Last"]} - {data["R#"]}"
    desti_fold = os.path.join(dep_path, department, fold_name)

    if os.path.exists(desti_fold):
        continue
    shutil.copytree(rids[data["R#"]], desti_fold)

# rids
```
